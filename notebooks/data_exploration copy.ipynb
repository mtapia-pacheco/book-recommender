```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("dylanjcastillo/7k-books-with-metadata")

print("Path to dataset files:", path)
```

```python
import pandas as pd

books = pd.read_csv(f"{path}/books.csv")
```

```python
books
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# plot heatmap of missing values
ax = plt.axes()
sns.heatmap(books.isna().transpose(), cbar=False, ax=ax, cmap="Blues")

plt.xlabel("Columns")
plt.ylabel("Missing values")

plt.show()
```

```python
# create numeric rep of missing books
import numpy as np

books["missing_description"] = np.where(books["description"].isna(), 1, 0)
books["age_of_book"] = 2026 - books["published_year"]
```

```python
# look at correlation matrix to look at potential bias
columns_of_interest = ["num_pages", "age_of_book", "missing_description", "average_rating"]

corr_mx = books[columns_of_interest].corr(method="spearman")  # non-continuous values

sns.set_theme(style="white")
plt.figure(figsize=(10, 10))
heatmap = sns.heatmap(corr_mx, annot=True, cmap="coolwarm",
                      cbar_kws={"label": "Spearman Correlation"})
heatmap.set_title("Correlation Heatmap")
plt.show()
```

```python
# missing description does not have a strong corr with these other vars -> we can drop the missing data
missing = books[books["description"].isna() |
      books["num_pages"].isna() |
      books["average_rating"].isna() |
      books["published_year"].isna()
]

missing
```

```python
miss_pct = missing.shape[0]/books.shape[0]
print(f'{miss_pct*100:.2f}% of the data is missing values.')
```

```python
books_complete = books[~(books["description"].isna()) &
      ~(books["num_pages"].isna()) &
      ~(books["average_rating"].isna()) &
      ~(books["published_year"].isna())
]

books_complete
```

```python
# look at distribution of categories
books_complete["categories"].value_counts().reset_index().sort_values("count", ascending=False)
```

```python
# check description length
books_complete["words_in_description"] = books_complete["description"].str.split().str.len()
```

```python
# set 25 words as a lower bound
books_complete_25 = books_complete[books_complete["words_in_description"] >= 25]
books_complete_25
```

```python
# combine title and subtitle if subtitle is available (more info)
books_complete_25["title_and_subtitle"] = (
    np.where(books_complete_25["subtitle"].isna(), books_complete_25["title"],
             books_complete_25[["title", "subtitle"]].astype(str).agg(": ".join, axis=1))
)

books_complete_25
```

```python
# tag each desc with isbn13 for later
books_complete_25["tagged_description"] = books_complete_25[["isbn13", "description"]].astype(str).agg(" ".join, axis=1)

books_complete_25
```

```python
# save new csv, extract only needed columns
(
    books_complete_25
    .drop(["subtitle", "missing_description", "age_of_book", "words_in_description"], axis=1)
    .to_csv("books_cleaned.csv", index=False)
)
```
